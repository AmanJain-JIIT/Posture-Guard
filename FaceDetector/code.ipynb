```python
import cv2

# Load the DNN based face detection model
prototxt_path = "A:\WORK\MajorProject\Bad-Posture-Detection-Alarming-System\FaceDetector\Models\deploy.prototxt"
model_path = r"A:\WORK\MajorProject\Bad-Posture-Detection-Alarming-System\FaceDetector\Models\res10_300x300_ssd_iter_140000_fp16.caffemodel"

face_model = cv2.dnn.readNetFromCaffe(prototxt_path, model_path)

# Create a VideoCapture object to capture frames from the webcam (change 0 to the video file path if using a file)
cap = cv2.VideoCapture(0)

# Check if the camera opened successfully
if not cap.isOpened():
    print("Error: Failed to open camera.")
    exit()

# Loop to capture frames and perform face detection
while True:
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Check if the frame was captured successfully
    if not ret:
        print("Error: Failed to capture frame.")
        break

    # Resize the frame to 300x300 pixels (required by the face detection model)
    blob = cv2.dnn.blobFromImage(frame, 1.0, (300, 300), (104.0, 177.0, 123.0))

    # Pass the blob through the face detection model
    face_model.setInput(blob)
    detections = face_model.forward()

    # Initialize a flag to indicate if a face is detected
    face_detected = False

    # Loop over the detections
    for i in range(detections.shape[2]):
        confidence = detections[0, 0, i, 2]

        # If confidence is above a certain threshold, consider it as a face
        if confidence > 0.5:
            face_detected = True
            break

    # Display "Face present" or "Face not present" directly on the frame
    text = "Face present" if face_detected else "Face not present"
    color = (0, 255, 0) if face_detected else (0, 0, 255)
    cv2.putText(frame, text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, color, 2)

    # Display the output
    cv2.imshow('Face Detection', frame)

    # Wait for 10 milliseconds
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

# Release the VideoCapture object and close all windows
cap.release()
cv2.destroyAllWindows()
```
